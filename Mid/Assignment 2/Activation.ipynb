**IMPLEMENTATION OF DIFFERENT ACTIVATION FUNCTION**


In order to add non-linearity to a neural network model, activation functions are mathematical operations that are applied to the outputs of the hidden layers. To function well, deep learning needs a lot of data. The output of a neural network model is determined by mathematical equations called activation functions. They support the network's usage of crucial data while reducing noise.

```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 2)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

 **1. Step Activation Function**

 In the perceptron network, step activation is used. When converting to a binary (0 or 1) or bipolar output, this is typically employed in single-layer networks (-1 or 1). Bipolar step function and Binary step function, respectively, are the names of these. The output in this situation is 1 (a neuron will fire) if the input value is larger than a threshold value, otherwise it is 0 or -1 in the case of a bipolar step function (neuron will not fire).







```python
y = np.zeros(len(x))
y[x>=0] = 1

plot_graph(y, "Step Activation Function")
```

**2. Sigmoid Activation Function**

The perceptron uses a highly strict logic known as the step activation function, which states that if a value is higher than a threshold value, the output will be 1 (indicating that a neuron will fire), otherwise 0 (indicating that a neuron won't fire).



```python

y = 1/(1 + np.exp(-x))

plot_graph(y, "Sigmoid Activation Function")
```

**3. Tanh Activation Function**

Tangent hyperbolic is signified by the acronym tanh. The activation function tanh is nonlinear. In multilayer neural networks, specifically for hidden layers, it is mostly utilized as an exponential function.




```python
y = np.tanh(x)

plot_graph(y, "Tanh Activation Function")
```

**4. ReLU Activation Function**

In the field of deep learning, the most popular activation function is known as Relu or Rectified Linear Activation Function. Relu delivers cutting-edge results while being extremely computationally effective.



```python
y = np.maximum(0, x)

plot_graph(y, "ReLU Activation Function")
```

 **5. ELU Activation Function**

 Exponential Linear Unit, sometimes known as ELU, is a function that tends to converge costs to zero more quickly and yield more precise results. ELU has a second alpha constant, which is positive, unlike other activation functions.





```python
alpha = 1.0

y = np.where(x>=0, x, alpha*(np.exp(x)-1))

plot_graph(y, "ELU Activation Function")
```

**6. SELU Activation Function**

The activation functions known as SELUs, or Scaled Exponential Linear Units, cause self-normalization. Neuronal activations in the SELU network automatically converge to zero mean and unit variance.





```python
def selu(x, scale=1.0507, alpha=1.6733):
    return np.where(x > 0, scale*x, scale*(alpha*np.exp(x)-alpha))

plot_graph(selu(x), "SELU Activation Function")
```
